```python
model_path = os.path.join('.', 'runs', 'classify', 'train8', 'weights', 'last.pt')

# Load a model
model = YOLO(model_path)  # load a custom model

# test the model on the images in my_test folder
# my_test folder has 2 folders inside it, Cat and Dog

true_positives = 0
true_negatives = 0
false_positives = 0
false_negatives = 0

# iterate over the images in my_test/Cat
for root, dirs, files in os.walk('my_test\Cat'):
    for file in files:
        test_image_path = os.path.join(root, file)
        try:
            results = model.predict(source=test_image_path)
            pred = results[0].probs.data.argmax()
            if pred == 0:
                true_negatives += 1
            else:
                false_positives += 1
        except:
            print("The image", file, "is corrupted")
for root, dirs, files in os.walk('my_test\Dog'):
    for file in files:
        test_image_path = os.path.join(root, file)
        try:
            results = model.predict(source=test_image_path)
            pred = results[0].probs.data.argmax()
            if pred == 1:
                true_positives += 1
            else:
                false_negatives += 1
        except:
            print("The image", file, "is corrupted")
print("True Positives:", true_positives)
print("True Negatives:", true_negatives)
print("False Positives:", false_positives)
print("False Negatives:", false_negatives)

# print the accuracy
print("Accuracy:", (true_positives + true_negatives)/(true_positives + true_negatives + false_positives + false_negatives))

# print the precision
print("Precision:", true_positives/(true_positives + false_positives))

# print the recall
print("Recall:", true_positives/(true_positives + false_negatives))

# print the F1 score
print("F1 Score:", 2*true_positives/(2*true_positives + false_positives + false_negatives))
```
